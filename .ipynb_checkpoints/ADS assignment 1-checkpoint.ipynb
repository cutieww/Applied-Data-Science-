```python
#function for generate 6 TCs
def generateTC2(av,iv,du):
    tc = []
    count_1 = du
    count_0 = iv-du
    if (av ==0):
        count = count_1 
        value = 1
    else:
        count = av
        value = 0
    for i in range(0,240):
        count = count-1
        tc.append(value) 
        if (count==0):
            if (value == 0):
                value = 1
                count = count_1
            else:
                value = 0
                count = count_0
        
    return tc
```

```python
#generate 6 tcs 
AV = [0,20,0,0,0,0]
IV = [30,45,60,40,40,40]
duration = [15,20,25,15,20,25]
first_tc1 = generateTC2(AV[0],IV[0],duration[0])
second_tc1 = generateTC2(AV[1],IV[1],duration[1])
third_tc1 = generateTC2(AV[2],IV[2],duration[2])
fourth_tc1 = generateTC2(AV[3],IV[3],duration[3])
fifth_tc1 = generateTC2(AV[4],IV[4],duration[4])
sixth_tc1 = generateTC2(AV[5],IV[5],duration[5])
```

```python
#generatee all the sms
def generateSM(x,start_v,end_v,start_h,end_h):
    while (start_v<=end_v):
        x[start_v][start_h:end_h] = 1
        start_v=start_v+1
    return x
```
